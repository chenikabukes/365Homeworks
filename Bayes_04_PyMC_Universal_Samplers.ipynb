### Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import pymc as pm
import numpy as np
from scipy import stats
import arviz as az
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
n = 100
true_theta, true_tau = 0, 1
data = stats.norm(loc=true_theta, scale=true_tau**(-0.5)).rvs(size=n)

with pm.Model() as model_normal_gamma:
    # Priors: Normal for theta and Gamma for tau (precision)
    theta = pm.Normal("theta", mu=0, sigma=1)
    tau = pm.Gamma("tau", alpha=0.5, beta=0.5)

    # Likelihood
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=data)

    # Sample from the posterior
    idata_ng = pm.sample(draws=2000, chains=2, tune=1000, target_accept=0.9)
```

```python
az.plot_trace(idata_ng)
plt.tight_layout()
```

- Both $\tau$ and $\theta$ have smooth unimodal posterior densities.
- The peak for $\tau$ is around $1.2$
- The peak of $\theta$ is very slightly below the true value of $0$. All of their respective chains are similar to each other.
- Since the time series fluctuates around stable means for both $\tau$ and $\theta$ it appears the sampler has converged and is mixing well.

```python
az.summary(idata_ng, round_to=5, hdi_prob=0.95)
```

- The estimated value of $\theta$ is decently close to its actual value of $0$.

- The estimate of $\tau$ is also close to its actual value of $1$.

- The true values of $\theta$ and $\tau$ are contained in the 95% hdi.

- The mcse is quite small for both $\theta$ and $\tau$ so the estimate is quite precise.

- The effective sample size is around $3000$ which is relatively close to $4000$.

```python
with pm.Model() as model_laplace_halfnorm:
    # Laplace prior for theta (non-normal)
    theta = pm.Laplace("theta", mu=0, b=1)

    # HalfNormal prior for sigma (standard deviation), so that tau = 1/sigma^2 is not Gamma
    sigma = pm.HalfNormal("sigma", sigma=1)

    # Likelihood written in terms of sigma (instead of tau)
    x_obs = pm.Normal("x_obs", mu=theta, sigma=sigma, observed=data)

    # Sample from the posterior
    idata_lh = pm.sample(draws=2000, chains=2, tune=1000, target_accept=0.9)

```

```python
az.plot_trace(idata_lh)
plt.tight_layout()
```

- Both $\theta$ and $\sigma$ have smooth unimodal posterior densities.

```python
az.summary(idata_lh, round_to=2)
```

- The estimated value of $\theta$ is decently close to its actual value of $0$.

- The estimate of $\sigma$ is also close to its actual value of $1$.

- The true values of $\theta$ and $\sigma$ are contained in the 95% hdi.

- The mcse is 0 for both $\theta$ and $\sigma$ so the estimate is quite precise.

- The effective sample size is around $4000$ which is close to or value.

```python
with pm.Model() as model_cauchy_halfcauchy:
    # Cauchy prior for theta (non-normal, heavy-tailed)
    theta = pm.Cauchy("theta", alpha=0, beta=1)

    # HalfCauchy prior for sigma (again, working with sigma)
    sigma = pm.HalfCauchy("sigma", beta=1)

    # Likelihood written in terms of sigma
    x_obs = pm.Normal("x_obs", mu=theta, sigma=sigma, observed=data)

    # Sample from the posterior
    idata_ch = pm.sample(draws=2000, chains=2, tune=1000, target_accept=0.9)

```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```



### How the Provided Slice Sampling Algorithm Works

1. **Choosing the Auxiliary Height $y $:**  
   For a given current value $ x $ (from the Markov chain) and an unnormalized density function $ f(x)$ (which we only know up to a constant), the algorithm begins by choosing an auxiliary value  

   $
   y \sim \text{Uniform}(0, f(x))
   $

   This auxiliary value $ y $ sets a horizontal level. Conceptually, you can imagine the graph of $ f(x) $ and a horizontal line at height $ y $. The idea is to sample uniformly from the region (or "slice") where the density exceeds this level.

2. **Finding the Slice  The Interval Where $ f(x) > y $:**  
   The code first defines a grid over the domain (using, for example, `x_grid = np.linspace(0,1,51)` or a finer grid). It then identifies the set of grid points for which

   $
   f(x_{\text{grid}}) > y
   $

   From these grid points, the leftmost and rightmost values (say, $ a $ and $ b $) are taken as initial endpoints. To be safe, the code extends the interval by a small amount (adding/subtracting one grid spacing) so that the endpoints are below the level $ y $ ($ f(a) < y $ and $ f(b) < y $).

3. **Proposing a New Sample from the Slice:**  
   A candidate value $x'$ is then proposed uniformly from the interval $[a, b].
   This is the first attempt to pick a new $ x $ that lies in the slice of the density.

4. **Acceptance Test and Interval Shrinking:**  
   - **If $ f(x') > y $:**  
     The candidate is accepted immediately because it lies under the curve (i.e. within the slice).
     
   - **If $ f(x') \le y $:**  
     The candidate is rejected. Moreover, depending on whether $ x' $ is to the left or right of the current $ x $, the algorithm shrinks the interval:
       - If $ x' < x $, then the new left endpoint becomes $ x' $ (i.e. we know that the region to the left of $ x' $ is no longer acceptable).
       - If $ x' \ge x $, then the new right endpoint becomes $ x' $.
     
     With this new interval, the algorithm recursively attempts to sample a candidate that satisfies $ f(x') > y $ by calling a helper function.

5. **Returning the New Sample:**  
   Once a candidate is found for which $ f(x') > y $, that value is returned (along with a counter of how many iterations were required). This new value becomes the next state in the Markov chain.

In summary, the algorithm slices the area under the unnormalized density $ f(x) $ at the random level $ y $ and then uniformly picks a new point from the resulting interval. If the candidate falls outside the desired region, the interval is shrunk, and the process repeats.

### Using Slice Sampling in a MetropolisWithinGibbs Framework

When implementing a Gibbs sampler, we need to update each parameter by sampling from its full conditional distribution. Often, the full conditionals are known only up to a normalizing constantthat is, we know a function proportional to the full conditional, but we cannot easily sample from it directly.

**Heres how slice sampling can replace a MetropolisHastings step for such a parameter:**

1. **Identify the Full Conditional:**  
   Suppose for a parameter $ \theta $ the full conditional density is given by  
   $
   f(\theta) \propto \text{(likelihood)} \times \text{(prior)}
   $
   Even if the normalizing constant is unknown, we can evaluate $ f(\theta) $ for any given $ \theta$.

2. **Set the Current Value as the Initial Point:**  
   In the Gibbs sampler, you already have a current value $ \theta^{(t)} $. This value is used as the starting point for the slice sampler.

3. **Generate an Auxiliary Variable $ y $:**  
   Compute the density at the current value, $ f(\theta^{(t)}) $, and draw  

   $
   y \sim \text{Uniform}(0, f(\theta^{(t)}))
   $

   This auxiliary variable defines a horizontal line across the density curve.

4. **Define the Slice:**  
   The slice is the set of all $ \theta $ such that  
   $
   f(\theta) > y
   $
   In the code, a grid (or an intervalexpansion method) is used to identify an interval $[a, b]$ where this condition holds.

5. **Draw a Candidate from the Slice:**  
   Sample a candidate $ \theta' $ uniformly from the interval $[a, b]$. Check if $ f(\theta') > y $:
   - **If Yes:** Accept $ \theta' $ as the new draw.
   - **If No:** Shrink the interval (remove the portion where $ f(\theta) $ is too low) and sample again.

6. **Integrate into the Gibbs Sampler:**  
   Replace the current $\theta$ with the accepted $ \theta' $ and continue to update other parameters in turn. This is effectively a MetropoliswithinGibbs update where the proposal is automatically adjusted by the slice sampling step.

**Clarifications:**

- **The Curve Being Sampled Beneath:**  
  The curve is the unnormalized full conditional density $ f(\theta) $ of the parameter. We are sampling uniformly from the region under this curve.

- **The Initial Value:**  
  The starting point for the slice sampling step is the current value $ \theta^{(t)} $ in the Gibbs chain. Its density $ f(\theta^{(t)}) $ is used to determine the height $ y $.

- **Steps to Create the Draw:**  
  1. Compute $ y \sim \text{Uniform}(0, f(\theta^{(t)})) $.  
  2. Determine an interval $[a, b]$ such that for all $ \theta \in [a, b] $, $ f(\theta) > y $.  
  3. Draw $ \theta' $ uniformly from $[a, b]$.  
  4. If $f(\theta') > y $, accept $\theta'$; otherwise, shrink the interval and repeat.

This procedure requires no tuning of a proposal distribution and adapts naturally to the shape of the full conditional density.
